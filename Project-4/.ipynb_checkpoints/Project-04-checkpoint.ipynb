```python
#CSV OPERATIONS
import csv
def csv_ops(lst=[],ops='read',name='jobs_draft.csv'):
    if(ops=='read'):
        with open(name,'rtb') as a:
            return [c for b in csv.reader(a) for c in b] #Flatten List
    elif(len(lst)):
        with open(name,'wtb') as a:
            wr=csv.writer(a)
            wr.writerow(lst)
            print 'SUCCESS!! Write',name
```

```python
#FILTER NON-ASCII CHARACTERS
def is_ascii(chars):
    if(len(chars)):
        #return ''.join([c for c in words if(ord(c)<128)]) #Simple Version
        out=''
        for c in chars:
            code=ord(c)
            if(code>31) and (code<127): #If approved ASCII
                if(code<48) or (code>122) or ((code>57) and (code<65)) or ((code>90) and (code<97)): c=' ' #Replace w <space>
                out+=c

        return ' '.join(out.split()) #Remove all whitespace characters (space, tab, newline, return, formfeed)
    else: return '' #Return Blank if blank
```

```python
#SPLIT JOINED WORDS
import re
from collections import Counter

def viterbi_segment(text):
    probs, lasts = [1.0], [0]
    for i in range(1, len(text) + 1):
        prob_k, k = max((probs[j] * word_prob(text[j:i]), j)
                        for j in range(max(0, i - max_word_length), i))
        probs.append(prob_k)
        lasts.append(k)
    words = []
    i = len(text)
    while 0 < i:
        words.append(text[lasts[i]:i])
        i = lasts[i]
    words.reverse()
    return words, probs[-1]

def word_prob(word): return dictionary[word] / total
def words(text): return re.findall('[a-z]+', text.lower()) 
dictionary = Counter(words(open('big.txt').read()))
max_word_length = max(map(len, dictionary))
total = float(sum(dictionary.values()))

viterbi_segment('responsibilitiesimplementation') #Example
```

```python
#WEBSCRAPING LIBRARIES
import os
import string
#import urllib2
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

site='https://jobscentral.com.sg'
chromedvr='/Users/davekho/Downloads/chromedriver' #Enter Chrome Driver Location
os.environ["webdriver.chrome.driver"]=chromedvr
dvr=webdriver.Chrome(chromedvr)
```

```python
#WEBSITE LOGIN
def web_login(url,usrname,pw,dvr):
    dvr.get(url)
    dvr.find_element_by_id('inputEmail').send_keys(usrname)
    dvr.find_element_by_id('inputPassword').send_keys(pw)
    dvr.find_element_by_id('submitBtn').click()
    WebDriverWait(dvr,5).until(EC.text_to_be_present_in_element((By.XPATH,'//li[@class="nav-item d-none d-xl-block"]/a[1]'),'Log Out')) #Explicit Wait in sec
    return 'Login Success'
```

```python
#GOTO WEBPAGE & RETURN "BEUTIFULSOUP" HTML
def go_page(url,dvr):
    print url
    resp=requests.get(url).status_code #Get Status Code
    #resp=urllib2.urlopen(url+link).getcode()
    if(resp==200): #If Status 200
        dvr.get(url)
        dvr.implicitly_wait(2) # Implicit Wait in sec
        html=dvr.page_source
        if(len(html)>100): #return BeautifulSoup(html,'lxml').body #Standard

            #jobscentral Customization
            print html[510:600]
            sp=BeautifulSoup(html,'lxml').body.find('div',{'class':'main-body'})
            if(sp==None):
                print 'ERROR : Content Missing'
                return None
            else: return sp

        else: #If No html content
            print 'NO HTML Content'
            return None
    else: #If Status =/= 200
        print 'STATUS CODE : '+resp
        return None
```

```python
#GET JOB LIST : Title & Link
def jobs_list(title):
    title.replace(' ','%20')
    prefix='/jobsearch?q='
    suffix='&pg='
    out,pay=[],[]
    for a in range(1,100):
        url=site+prefix+title+suffix+str(a)
        sp=go_page(url,dvr)
        if(sp==None): #If No html content
            print a,':','Content Missing'
            break

        #Check for content section
        main=sp.find('div',{'class':'col-md-8 col-lg-9'}) #Content Section
        if(main==None):
            print a,': No Main Content'
            break

        #Check for Duplicate/Relevant
        #Get Job title & link
        elem=main.find_all('div',{'class':'col col-lg-9'})
        for b in elem:
            for c in b.findChildren('a'):
                tmp=[c.text.strip(),c['href']] #[Title,Link]
                if('/job/' in tmp[1]): #If Relevant Title & Link not in list & is Right link
                    out+=[tmp[0]+'|'+tmp[1]] #Get Title & Link

        #Get Pay
        elem=sp.find_all('div',{'_ngcontent-c11':'','class':'pay'})
        for a in elem: pay.append(a.text.strip())

        #Find "NEXT" Btn
        btn_next=main.find('a',{'aria-label':'Next','class':'page-link'},href=True)
        #print a,':',btn_next.parent['class']
        if(btn_next==None) or (btn_next['href']==url) or (btn_next.parent['class']!=['page-item']): #"Next" btn: If No btn OR btn url = current url OR wrong DOM parent class
            print a,': DONE' #If Got 'Next' btns
            break

    print title,':',len(out),' records\n'
    return out
```

<p>
The Job Details page that contains the job description has several layout formats. the code below has been modified several times to accomodate the different formats.
</p>

```python
#GET JOB DETAILS
def job_details(url):
    #Goto webpage
    sp=go_page(url,dvr)
    WebDriverWait(dvr,5).until(EC.text_to_be_present_in_element((By.TAG_NAME,'h3'),'Job Description')) #Explicit Wait in sec

    #Check for content section
    main=sp.find('div',{'class':'container p-0'}) #Content Section
    if(main): #If Got Content Section

        #Compare Title
        title=main.find('h1',{'_ngcontent-c3':''})
        title=title.text.strip() if title else ''
            
        #Get Company Name
        co=main.find('span',{'class':'d-block d-sm-inline'})
        co=co.text.strip() if(co) else ''

        #Get Job Snapshot
        pay=''
        snapshot=[]
        snap=main.find('job-snapshot')
        if(snap):
            snap=snap.find_all('div',{'_ngcontent-c10':'','class':'row'})
            for a in snap:
                for b in a.children:
                    if(b.name=='div'): snapshot+=[b.text.strip().replace('\n','')]
            snapshot=np.reshape(snapshot,(-1,2))

            #Get Salary from Snapshot
            for a in snapshot:
                if 'Salary' in a[0]:
                    pay=a[1] #Set Pay
                    print '1. pay',pay

                    #CONVERT PAY TO MONTHLY
                    a[1]=a[1].lower()
                    tmp=a[1].split(' ')
                    pay=[]
                    for b in tmp:
                        try:
                            salary=int(b.replace(',',''))
                            if(salary<1000): continue #Skip if <1000
                            elif(salary>10000) and ('year' in a[1]) or ('annual' in a[1]) or ('annum' in a[1]): salary/=12 #Get Monthly Salary
                            pay.append(salary)
                        except: pass

                    #Check For Salary Range
                    print '2. pay',pay
                    tmp=len(pay)
                    if(tmp==1): pay=str(pay[0])
                    elif(tmp==2):
                        #pay=str((pay[0]+pay[1])/2) #Get Salary Mean
                        pay=str(pay[0])+' - '+str(pay[1]) #Get Range
                    break

        #Get Job Description
        describe=str()
        des=main.find('h3',{'_ngcontent-c3':''})
        if(des) and (des.text=='Job Description'):
            des=des.next_sibling.next_sibling
            if(des.name=='div'):
                if(des.text): #If "div" got text
                    describe+=des.text.strip()#.decode(encoding='utf-32')
                    if(describe[-1]!='.'): describe+='.' #Add "." if none
                for a in des.children:
                    tmp=a.name
                    if(tmp=='p'): #If paragraph
                        try: describe+=tmp.text.strip()
                        except: describe+=tmp.strip()
                        if(describe[-1]!='.'): describe+='.' #Add "." if none
                    elif(tmp=='ul'): #if unordered list
                        for b in a.children:
                            if(b.name=='li'): describe+=b.text.strip()
                            if(describe[-1]!='.'): describe+='.' #Add "." if none

            #Replace string contents
            describe=describe.replace("\n",".").replace(',',' ').replace('&nbsp','') #Replace string
            #describe=describe.replace(u"\xa0",u"").replace(u"\u2013",u"-").replace(u"\u2019",u"").replace(u'\u2022',u'').replace(u'.p.p','') #Replace UTF-8
            describe=describe.map(lambda a: is_ascii(a)) #Filter Non-ASCII

        else: describe=''

    else: return title,': Content Missing'
    return [co,describe,title,pay]
```

<p>
The salary column is a string with words. Most of the salaries are a range. They also come in yearly sum or monthly. we need to reformat the standardize the salary column to monthly mean salary.
</p>

```python
#REPLACE "SALARY" VALUES TO PREPARE DATA TO COMPUTE
salary_replace=[[u'\xa0',u''],['to',' - '],['k','000'],['annually',' per annum'],[','],['$'],['min'],['sgd'],['contract'],['monthly'],
                ['1 <span>vacancy</span>'],['2 <span>vacancies</span>'],['4 <span>vacancies</span>'],['6 <span>vacancies</span>']]

for a in salary_replace:
    if(len(a)==1): df.salary = df['salary'].str.replace(a[0],'')
    else: df.salary = df.salary.str.replace(a[0],a[1])
```

<p>Found out that the salaries for "achieve" located in the description and not salary column. we need to do some processing to revert the error and get an average of their ranges. 
</p>

```python
#GET "ACHIEVE" SALARY
ref=df[(df.company=='achieve career consultant pte ltd (recruitment firm)')].index.values
for a in ref:
    pay=[]
    desc=df.loc[a,'description'].replace('$ ','$').split(' ')
    for c in desc:
        lon=len(pay)
        if(lon==2): break
        elif('$' in c):
            try:
                c=int(c[1:].replace(',','').replace('.','').replace('k','000').encode('ascii'))
                pay.append(c)
            except: continue
    if(lon==2): df.loc[a,'salary']=(pay[0]+pay[1])/2
    elif(lon==1): df.loc[a,'salary']=pay[0]
```

```python
#CONVERT "SALARY" RANGE TO MEAN
for a in range(df['salary'].shape[0]):
    pay=[]
    string=str(df.loc[a,'salary'])
    if(string) and (any(b.isdigit() for b in string)==True): #If GOT 'salary' value & digit in string
        tmp=string.replace('sgd','').strip().split(' ') #Stringfy, replace sgd, & split
        for b in tmp:
            try:
                salary=int(b.replace(',',''))
                if('year' in string) or ('annual' in string) or ('annum' in string): salary/=12 # Yearly ==> Monthly
                pay.append(salary)
            except: continue

        tmp=len(pay) #n of salaries?
        df.loc[a,'salary']=pay[0] if(tmp==1) else ((pay[0]+pay[1])/2) if(tmp>1) else np.nan #Mean If Got salary range?
        if df.loc[a,'salary']<100: df.loc[a,'salary']=np.nan
    else: df.loc[a,'salary']=np.nan #If Empty or NO digit in string
    #print a,':',tmp,pay,'==>',df.iloc[a,3]

df.salary=pd.to_numeric(df.salary)
#df.salary.value_counts()
#df.salary.unique()
```

```python
#TO SHOW MODEL MOST USED FEATURES
def show_top_features(limit,classes,features,coefs):
    if(type(features)!='numpy.ndarray'): features=np.array(features) #Convert if Not Numpy array
    if len(classes) > 2:
        print 'Top',limit,'Features:-'
        for a,b in enumerate(classes):
            top=np.argsort(coefs[a])[-limit:] #Get Top 10 Indexes
            print '%s : %s' % (b,' , '.join(features[top]))
    else:
        top=np.argsort(coefs[0])[-limit:] #Get Top 10 Indexes
        print 'Top '+str(limit)+' features found in %s:' % (classes[1])
        print '%s : ' % (' , '.join(features[top]))
```

```python
#LOGISTIC REGRESSION & MULTINOMIAL NAIVE BAYES
from sklearn import metrics
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score,cross_val_predict

def bayes_logistic(vect,cols,X_train,X_test,y_train,y_test):
    X_train_vec=vect.transform(X_train)
    X_test_vec=vect.transform(X_test)
    print 'For',X_train_vec.shape[1],'Features\n'

    for a in range(2):
        #Create Model Instance
        model=MultinomialNB() if a else LogisticRegression(random_state=42)
        print model.__module__

        #Cross-Validate
        scores=cross_val_score(model,X_train_vec,y_train,cv=5)
        print 'Cross-Val Score: {:.3f} ({:.3f})\n'.format(scores.mean(),scores.std()*2)

        #Fit Data
        model.fit(X_train_vec,y_train)

        #Show Top Features
        classes=model.classes_
        coefs=model.coef_
        show_top_features(10,classes,cols,coefs)

        #Predict & Score
        y_pred=model.predict(X_test_vec)
        print '\nAccuracy: ',metrics.accuracy_score(y_test,y_pred),'\n'
        print 'Classification Report:\n',metrics.classification_report(y_test,y_pred),'\n\n'
```

<p>
We decide the employ 2 methods to count the frequencies of all the words in the job descriptions; CountVectorizer giving equal weight to each word and tfidfvectorizer giving varying weights to different words. For both approaches, we also used a simple dict to ignore commonly used words without much meaning such as 'are' and is.
For the purpose of modeling, we decide to use a very simple classification approach "Multinomial logistic regression" as a basline model. We further use Multinomial Naive Bayes as it can make strong independence assumptions rather than the particular distribution of each feature. its origins is from Bayes Theorem. Naive Bayes model assumes that each of the features it uses are conditionally independent of one another given some class. Supposedly it works well for data which can easily be turned into counts, eg. word counts.

```python
#SPELL CHECKER 2
import re
from collections import Counter

def words(text): return re.findall(r'\w+', text.lower())
WORDS = Counter(words(open('big.txt').read()))
def P(word, N=sum(WORDS.values())): return WORDS[word] / N #Probability of word
def correction(word): return max(candidates(word), key=P) #Most probable spelling correction for word
def candidates(word): return (known([word]) or known(edits1(word)) or known(edits2(word)) or [word]) #Generate possible spelling corrections for word
def known(words): return set(w for w in words if w in WORDS) #The subset of `words` that appear in the dictionary of WORDS

def edits1(word): #All edits that are one edit away from word
    letters    = 'abcdefghijklmnopqrstuvwxyz'
    splits     = [(word[:i], word[i:])    for i in range(len(word) + 1)]
    deletes    = [L + R[1:]               for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1]
    replaces   = [L + c + R[1:]           for L, R in splits if R for c in letters]
    inserts    = [L + c + R               for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)

def edits2(word): return (e2 for e1 in edits1(word) for e2 in edits1(e1)) #All edits that are two edits away from word

print correction('responsibilitiesbusiness') #Example
```
