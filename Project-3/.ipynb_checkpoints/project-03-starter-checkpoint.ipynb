```python
#COUNT UNIQUE,NULL,ZERO OF ALL COLUMNS
def count_weird(df):
    df_info=pd.DataFrame()
    rows=float(df.shape[0])
    print int(rows),'rows &',df.shape[1],'columns'
    for a,b in df.items():
        msg,val=[],[str(df[a].dtype),df[a].nunique(),df[a].isnull().sum(),0] #Count [Unique,null]
        if(val[1]!=rows): msg.append(str(val[1])+' Unique')
        if(val[2]): msg.append(str(val[2])+' Null {0:.1f}%'.format(val[2]/rows*100))
        if(val[0]=='object'): #String / Object
            wierd=list(df[a].str.match(r'^(~|`|!|@|#|\$|%|\^|&|\*|\(|\)|-|_|\+|=|\{|\}|\[|\]|\\|\||:|;|"|<|>|,|\.|\?|/|\s)+$')) #Match Cell with weird input
            if True in wierd: print 'WIERD '+a+' :',[str(c)+': '+str(wierd[c]) for c in range(len(wierd)) if wierd[c] is True] #If Got Match
        else: #Other Dtypes
            val[3]=df[df[a]==0].shape[0]
            if(val[3]>0):
                #df[a].replace(0,np.nan,inplace=True) #OPTIONAL: 0 ==> NaN
                msg.append(str(val[3])+' Zero {0:.1f}%'.format(val[3]/rows*100))
        df_info=df_info.append({'Column':str(a),'Type':val[0],'Unique':val[1],'Null':val[2],'Null%':round(val[2]/rows*100,1),'Zero':val[3],'Zero%':round(val[3]/rows*100,1)},ignore_index=True)
        #if(len(msg)>1): print a+' : '+' | '.join(msg)
    return df_info.sort_values('Null',ascending=False)


#CHECK IF RELATED FEATUREs NAN
def check_nan(df,col,compare):
    df_all=df.loc[(df[col].isnull()),compare] #Get Only NaN Columns
    diff=df_all.size-df_all.notna().size
    if diff: print(col,':',diff,'non-NaN')
    else: print(col,'OK')
```

```python
#TRANSFORM NUMERICAL FEATURE TO NORMAL DISTRIBUTION
def normalize_data(df,limit=.5,diff=.2):
    '''
    Function auto-select best transformation to normalize data (based on skewness)
    limit : Skew Limit
    diff : Min skew difference for transform
    '''
    name=['Log','Log +1','Log 10','Square Root','Cube Root','Power 2','Exponential','Exponential 2','Exponential -1']
    print 'SKEWNESS TRANSFORM'
    for col in numerical:
        orig=round(np.abs(df[col].skew()),2)
        if(orig>limit):
            skew=[] #List of Skewness after transform
            method=[np.log(df[col]),np.log1p(df[col]),np.log10(df[col]),np.sqrt(df[col]),np.cbrt(df[col]),np.power(df[col],2),np.exp(df[col]),np.exp2(df[col]),np.expm1(df[col])] #Methods to transform feature
            for a in method: #Calc skew for each method
                b=round(np.abs(a.skew()),2)
                skew.append(9999.) if np.isnan(b) else skew.append(b) #float if skew = NaN else Add skew
            mini=min(skew) #Get lowest skew
            if(orig-mini>diff):
                i=skew.index(mini)
                df[col]=method[i]
                print col,orig,' ==> ',mini,name[i],np.where(mini>limit,'skewed','') #Transformed Feature Summary
            #else: print col,orig,' =/= ',mini,name[i],np.where(mini>limit,'skewed','') #Un-transform Feature Summary
```

```python
from sklearn.model_selection import cross_val_score,cross_val_predict
from sklearn import metrics
```

```python
#LASSO METHOD
from sklearn.linear_model import Lasso,LassoCV
def use_lasso(X,y,cols,folds=5):
    #Find Optimal Alpha
    optimal=LassoCV(cv=folds,verbose=1)
    optimal.fit(X,y)
    print 'LASSO\nOptimal Alpha:',optimal.alpha_

    #Use Optimal Alpha
    model=Lasso(alpha=optimal.alpha_)
    scores=cross_val_score(model,X,y,cv=folds)
    score_avg=scores.mean()
    print 'Score: {:.3f} ({:.3f})\n'.format(score_avg,scores.std())

    #Get Top 10 Features
    model.fit(X,y)
    top=[[a,b] for a,b in zip(cols,model.coef_)]
    print pd.DataFrame(top,columns=['Feature','Coefficient']).sort_values(by='Coefficient',ascending=False)[:10]
    return model,score_avg
```

```python
#RIDGE METHOD
from sklearn.linear_model import Ridge,RidgeCV
def use_ridge(X,y,cols,folds=5):
    #Find Optimal Alpha
    optimal=RidgeCV(alphas=np.logspace(-2,7),cv=folds)
    optimal.fit(X,y)
    print 'RIDGE\nOptimal Alpha:',optimal.alpha_

    #Use Optimal Alpha
    model=Ridge(alpha=optimal.alpha_)
    scores=cross_val_score(model,X,y,cv=folds)
    score_avg=scores.mean()
    print 'Score: {:.3f} ({:.3f})\n'.format(score_avg,scores.std())

    #Get Top 10 Features
    model.fit(X,y)
    top=[[a,b] for a,b in zip(cols,model.coef_)]
    print pd.DataFrame(top,columns=['Feature','Coefficient']).sort_values(by='Coefficient',ascending=False)[:10]
    return model,score_avg
```

```python
#ELASTIC-NET METHOD
from sklearn.linear_model import ElasticNet,ElasticNetCV
def use_elastic(X,y,cols,folds=5):
    #Find Optimal Alpha & L1
    optimal=ElasticNetCV(cv=folds)
    optimal.fit(X,y)
    print 'ELASTIC-NET\nOptimal Alpha:',optimal.alpha_,'\nL1 Ratio:',optimal.l1_ratio_

    #Use Optimal Alpha & L1 Ratio
    model=ElasticNet(alpha=optimal.alpha_,l1_ratio=optimal.l1_ratio_)
    scores=cross_val_score(model,X,y,cv=folds)
    score_avg=scores.mean()
    print 'Score: {:.3f} ({:.3f})\n'.format(score_avg,scores.std())

    #Get Top 10 Features
    model.fit(X,y)
    top=[[a,b] for a,b in zip(cols,model.coef_)]
    print pd.DataFrame(top,columns=['Feature','Coefficient']).sort_values(by='Coefficient',ascending=False)[:10]
    return model,score_avg
```

```python
def use_lasso_ridge_elastic(X_train,y_train,X_test,y_test,cols,min_score=.7,folds=10):
    print X_train.shape[1],'features\n'
    for Model in [use_lasso,use_ridge,use_elastic]:
        print Model.__name__
        model,score=Model(X_train,y_train,cols,folds)

        if(score<min_score): 
            print '\nScore',score,'< Minimum',min_score
        else: #If Model Mean Score >= min_score
            y_pred=cross_val_predict(model,X_test,y_test,cv=folds)
            print '\ny-Predict\nAccuracy:',metrics.r2_score(y_test,y_pred),'\nMSE:',metrics.mean_squared_error(y_test,y_pred),'\n\n'
```

```python
from sklearn.utils import resample
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import SGDClassifier

def bootstrap_imbalance_classify(major,minor,X_test,y_test,iters=5):
    coefs=[]
    scores=[]
    for a in range(iters):
        #Up-Sample Minority & Combine
        minor_sampl=resample(minor,replace=True,n_samples=major.shape[0]) #Up-sample w Replacement
        X_train_sampl=pd.concat([major,minor_sampl]) #Combine majority w resampled minority

        #Double-Check Counts
        tmp=X_train_sampl['SaleCondition_Abnorml'].value_counts()
        if(tmp[0]-tmp[1]!=0):
            print a,': Majority count =/= Minority count'
            break

        #Remove "SaleCondition_Abnorml" from Data
        y_train_sampl=X_train_sampl.SaleCondition_Abnorml
        X_train_sampl.drop(columns=['SaleCondition_Abnorml'],inplace=True)

        #Hyperparameters
        params={
            'alpha': [1e-4,1e-3,1e-2,1e-1,1e0,1e1,1e2,1e3],
            'loss': ['log','hinge','modified_huber'],
            'penalty': ['l1'], #lasso
            #'l1_ratio': np.linspace(0,1,num=5), #0=L2 ==> 1=L1
            'max_iter': np.arange(5,15,2),
            'n_jobs': [-1]} #Use all CPUs

        clf=GridSearchCV(SGDClassifier(),params)
        clf.fit(X_train_sampl,y_train_sampl)
        coefs+=[clf.best_estimator_.coef_]
        scores+=[clf.best_score_]

        print 'ITERATION',a,'\nScore',clf.best_score_,'\nParameters:',clf.best_params_
        #print metrics.classification_report(y_test,clf.predict(X_test)),'\n',

    return coefs,scores
```
