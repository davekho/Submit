```python
# Load the data
df=pd.read_csv('./housing.csv')
#print df.info(memory_usage=True,null_counts=True)
df.describe(include='all')
```

```python
df=df[df.MSZoning.isin(['FV','RH','RL','RP','RM'])] #Filter to Residential Zone Only
df.MSSubClass=df.MSSubClass.astype(str) #Int ==> Str (categorical class)
df.drop(['Id','GarageCars'],axis=1,inplace=True) #Drop useless Id
df_ref=df.copy(deep=True) #Create reference copy
print df.isnull().sum().sum(),'null'
print df.info(memory_usage=True,null_counts=True)
```

```python
#GET NUMERICAL & CATEGORICAL FEATURES
type_cat=list(df.loc[:,(df.dtypes=='object')].columns.values)
type_num=list(df.loc[:,(df.dtypes!='object')].columns.values)
type_num.remove('SalePrice')
print type(type_num),type_num
```

```python
#GET FIX / RENOVATABLE FEATURES
col_fix=['OverallQual','OverallCond','RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','ExterQual','ExterCond', #Fixable features + 'SalePrice'
         'BsmtCond','BsmtFinType1','BsmtFinType2','Heating','HeatingQC','CentralAir','Electrical','KitchenQual',
         'Fireplaces','FireplaceQu','GarageFinish','GarageQual','GarageCond','PavedDrive','PoolQC','Fence']

col_unfix=[] #Unfixable features
for a in df.columns.values:
    if a not in col_fix: col_unfix.append(a)

print df.shape[1]-len(col_unfix),'=',len(col_fix)
```

```python
#COUNT UNIQUE,NULL,ZERO OF ALL COLUMNS
def count_weird(df):
    df_info=pd.DataFrame()
    rows=float(df.shape[0])
    print int(rows),'rows &',df.shape[1],'columns'
    for a,b in df.items():
        msg,val=[],[str(df[a].dtype),df[a].nunique(),df[a].isnull().sum(),0] #Count [Unique,null]
        if(val[1]!=rows): msg.append(str(val[1])+' Unique')
        if(val[2]): msg.append(str(val[2])+' Null {0:.1f}%'.format(val[2]/rows*100))
        if(val[0]=='object'): #String / Object
            wierd=list(df[a].str.match(r'^(~|`|!|@|#|\$|%|\^|&|\*|\(|\)|-|_|\+|=|\{|\}|\[|\]|\\|\||:|;|"|<|>|,|\.|\?|/|\s)+$')) #Match Cell with weird input
            if True in wierd: print 'WIERD '+a+' :',[str(c)+': '+str(wierd[c]) for c in range(len(wierd)) if wierd[c] is True] #If Got Match
        else: #Other Dtypes
            val[3]=df[df[a]==0].shape[0]
            if(val[3]>0):
                #df[a].replace(0,np.nan,inplace=True) #OPTIONAL: 0 ==> NaN
                msg.append(str(val[3])+' Zero {0:.1f}%'.format(val[3]/rows*100))
        df_info=df_info.append({'Column':str(a),'Type':val[0],'Unique':val[1],'Null':val[2],'Null%':round(val[2]/rows*100,1),'Zero':val[3],'Zero%':round(val[3]/rows*100,1)},ignore_index=True)
        #if(len(msg)>1): print a+' : '+' | '.join(msg)
    return df_info.sort_values('Null',ascending=False)


#CHECK IF RELATED FEATUREs NAN
def check_nan(df,col,compare):
    df_all=df.loc[(df[col].isnull()),compare] #Get Only NaN Columns
    diff=df_all.size-df_all.notna().size
    if(diff): print col,':',diff,'non-NaN'
    else: print col,'OK'
```

```python
count_weird(df)[:20]
```

```python
#CHECK IF RELATED FEATUREs NAN
check_nan(df,'GarageCond',('GarageQual','GarageType','GarageFinish','GarageYrBlt'))
check_nan(df,'BsmtExposure',('BsmtFinType2'))
check_nan(df,'BsmtQual',('BsmtCond','BsmtFinType1'))
check_nan(df,'MasVnrType',('MasVnrArea'))
```

```python
#ADD NEW FEATURES
df['Age']=df['YrSold']-df['YearBuilt']
df['TotalSF']=df['TotalBsmtSF']+df['1stFlrSF']+df['2ndFlrSF']
type_num.extend(['Age','TotalSF'])
col_unfix.extend(['Age','TotalSF'])

#Remove YearBuilt
col_unfix.remove('YearBuilt')
type_num.remove('YearBuilt')
type_num.extend(['Age','TotalSF'])
df.drop(['YearBuilt'],axis=1,inplace=True)
```

```python
#FILL NANs & REPLACE NONE
df.replace('None','NA',inplace=True) #Convert all "None" ==> "NA"

#Fill as NA
for col in ('PoolQC','MiscFeature','Alley','Fence','FireplaceQu','GarageCond','GarageQual','GarageType','GarageFinish','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','MasVnrType'):
    try:
        df[col].fillna('NA',inplace=True)
    except: pass

#Fill as 0
for col in ('GarageYrBlt','GarageArea','GarageCars','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','MasVnrArea'):
    try:
        df[col].fillna(0,inplace=True)
    except: pass

#Fill as Mode
df.Electrical.fillna(df.Electrical.mode()[0],inplace=True) 

#Fill as Median
df['LotFrontage']=df.groupby('Neighborhood')['LotFrontage'].transform(lambda a: a.fillna(a.median()))
df.GarageYrBlt=df.GarageYrBlt.astype(int) #float ==> Integer
print df.isnull().sum().sum(),'null'
```

```python
#CONVERT CATEGORICAL TO NUMERICAL RANKING (None ==> Good)
cols=[['Alley','Street'],
      ['BsmtFinType1','BsmtFinType2'],
      ['ExterQual','ExterCond','BsmtQual','BsmtCond','HeatingQC','KitchenQual','FireplaceQu','GarageQual','GarageCond','PoolQC']]

for a in range(len(cols)):
        if a==0:
            for col in cols[a]: df[col]=df[col].map({'NA':0,'Grvl':1,'Pave':2})
        elif a==1:
            for col in cols[a]: df[col]=df[col].map({'NA':0,'Unf':1,'LwQ':2,'Rec':3,'BLQ':4,'ALQ':5,'GLQ':6})
        else:
            for col in cols[a]: df[col]=df[col].map({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})

df.LotShape=df.LotShape.map({'NA':0,'IR3':1,'IR2':2,'IR1':3,'Reg':4})
df.Utilities=df.Utilities.map({'NA':0,'ELO':1,'NoSeWa':2,'NoSewr':3,'AllPub':4})
df.LandSlope=df.LandSlope.map({'NA':0,'Sev':1,'Mod':2,'Gtl':3})
df.BsmtExposure=df.BsmtExposure.map({'NA':0,'No':1,'Mn':2,'Av':3,'Gd':4})
df.Functional=df.Functional.map({'NA':0,'Sal':1,'Sev':2,'Maj2':3,'Maj1':4,'Mod':5,'Min2':6,'Min1':7,'Typ':8})
df.GarageFinish=df.GarageFinish.map({'NA':0,'Unf':1,'RFn':2,'Fin':3})
df.PavedDrive=df.PavedDrive.map({'N':1,'P':2,'Y':3})
df.Fence=df.Fence.map({'NA':0,'MnWw':1,'GdWo':2,'MnPrv':3,'GdPrv':4})

df.CentralAir=df.CentralAir.map({'N':0,'Y':1}) #Binarize

#Update numerical & categorial feature lists
tmp=[b for a in cols for b in a] #Flatten list
tmp+=['LotShape','Utilities','LandSlope','BsmtExposure','CentralAir','Functional','GarageFinish','PavedDrive','Fence']
type_num+=tmp
for a in tmp: type_cat.remove(a)
#print type_num
#df.info()
```

```python
#INVESTIGATE SalePrice
print 'skew',df.SalePrice.skew(),' | ','kurt',df.SalePrice.kurt(),'\n',df.SalePrice.describe()
sns.distplot(df.SalePrice,fit=stats.norm)
fig=plt.figure()
stats.probplot(df.SalePrice,plot=plt)
```

```python
#Normalize SalePrice
df.SalePrice=np.log(df.SalePrice)
print 'skew',df.SalePrice.skew(),' | ','kurt',df.SalePrice.kurt(),'\n',df.SalePrice.describe()
```

```python
#Check Nemerical Features Distribution
for col in type_num: print col,':',round(df[col].skew(),2)
```

```python
#CHECK & TRANSFORM NUMERICAL FEATURE DISTRIBUTION IF ABNORMAL
limit=.5
for col in type_num:
    orig=round(np.abs(df[col].skew()),2)
    if(orig>limit):
        method=[np.log(df[col]),np.log1p(df[col]),np.log10(df[col]),np.sqrt(df[col]),(df[col]**2)] #Methods to transform feature
        skew=[]
        for a in method:
            tmp=round(np.abs(a.skew()),2)
            if(np.isnan(tmp)): skew.append(9999.)
            else: skew.append(tmp)
        m=min(skew)
        if(orig>m):
            df[col]=method[skew.index(m)]
            print col,orig,' ==> ',m
        else: print col,orig,' =/= ',m
```

```python
#PLOT CORRELATION HEATMAP
fig,ax=plt.subplots(figsize=(22,22))
cor=df.corr()
sns.heatmap(cor,square=True)
plt.show()
```

```python
#Top 10 Features affecting SalesPrice
fig,ax=plt.subplots(figsize=(10,10))
cols=df.corr().nlargest(10,'SalePrice')['SalePrice'].index
corr=np.corrcoef(df[cols].values.T)
sns.heatmap(corr,cbar=True,annot=True,square=True,fmt='.2f',yticklabels=cols.values,xticklabels=cols.values,ax=ax)
plt.show()
```

```python
#CREATE CLASSIFICATION DUMMY COLUMNS
drop_last=[]
for a in type_cat: drop_last+=[a+'_'+str(df[a].unique()[-1])] #Get drop col names
df=pd.get_dummies(df,columns=type_cat) #Create dummy cols
df.drop(drop_last,axis=1,inplace=True)
#df.columns.values
```

```python
#ASSIGN TRAIN / TEST SETs
X_train=df[(df.YrSold<2010)]
y_train=X_train.SalePrice
X_train.drop(['SalePrice'],axis=1,inplace=True)

X_test=df[(df.YrSold==2010)]
y_test=X_test.SalePrice.values
X_test.drop(['SalePrice'],axis=1,inplace=True)

print 'Train:',X_train.shape,'  Test:',X_test.shape
```

```python
#STANDARDIZE FIX & UNFIXED FEATUREs
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
X_train[type_num]=scale.fit_transform(X_train[type_num])
X_test[type_num]=scale.transform(X_test[type_num])

print X_train.isnull().sum().sum(),X_test.isnull().sum().sum()
print 'Train:',X_train.shape,'  Test:',X_test.shape
```

```python
#GET UNFIXABLE DATASET
ref=[a for a in X_train.columns for b in col_unfix if a.find(b)==0] #Get unfixable dataset
X_pred=X_test[ref]
X=X_train[ref]
y=y_train
```

```python
from sklearn.linear_model import ElasticNetCV,ElasticNet,Lasso,LassoCV,Ridge,RidgeCV,LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score,cross_val_predict
from sklearn.metrics import mean_squared_error
from sklearn import metrics
folds=5
```

```python
optimal_lasso=LassoCV(cv=folds,verbose=1)
optimal_lasso.fit(X,y)
print optimal_lasso.alpha_
lasso=Lasso(alpha=optimal_lasso.alpha_)
score=cross_val_score(lasso,X,y,cv=folds)
print("\nLasso score: {:.3f} ({:.3f})\n".format(score.mean(),score.std()))

lasso.fit(X,y)
summary=[]
for a,b in zip(col_unfix,lasso.coef_): summary.append([a,b])
print pd.DataFrame(summary,columns=['Feature','Coefficient']).sort_values(by='Coefficient',ascending=False)[:10]

if(score.mean>.8):
    predict=cross_val_predict(lasso,X_pred,y_test,cv=folds)
    print 'Accuracy:',metrics.r2_score(y_test,predict)
    print 'MSE:',mean_squared_error(y_test,predict)
```

```python
ridge_alphas=np.logspace(-2,7,50)
optimal_ridge=RidgeCV(alphas=ridge_alphas,cv=folds)
optimal_ridge.fit(X,y)
print optimal_ridge.alpha_
ridge=Ridge(alpha=optimal_ridge.alpha_)
score=cross_val_score(ridge,X,y,cv=folds)
print("\nRidge score: {:.3f} ({:.3f})\n".format(score.mean(),score.std()))

ridge.fit(X,y)
summary=[]
for a,b in zip(col_unfix,ridge.coef_): summary.append([a,b])
print pd.DataFrame(summary,columns=['Feature','Coefficient']).sort_values(by='Coefficient',ascending=False)[:10]

if(score.mean>.8):
    predict=cross_val_predict(ridge,X_pred,y_test,cv=folds)
    print 'Accuracy:',metrics.r2_score(y_test,predict)
    print 'MSE:',mean_squared_error(y_test,predict)
```

```python
optimal_elastic=ElasticNetCV(cv=folds)
optimal_elastic.fit(X,y)
print optimal_elastic.alpha_
print optimal_elastic.l1_ratio_
elastic=ElasticNet(alpha=optimal_elastic.alpha_,l1_ratio=optimal_elastic.l1_ratio_)
score=cross_val_score(elastic,X,y,cv=folds)
print("\nElastic-Net score: {:.3f} ({:.3f})\n".format(score.mean(),score.std()))

elastic.fit(X,y)
summary=[]
for a,b in zip(col_unfix,elastic.coef_): summary.append([a,b])
print pd.DataFrame(summary,columns=['Feature','Coefficient']).sort_values(by='Coefficient',ascending=False)[:10]

if(score.mean>.8):
    predict=cross_val_predict(elastic,X_pred,y_test,cv=folds)
    print 'Accuracy:',metrics.r2_score(y_test,predict)
    print 'MSE:',mean_squared_error(y_test,predict)
```
